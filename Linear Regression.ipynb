```
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```
#Creating a class for Linear Model
import torch.nn as nn
```

```
class Linear_reg(nn.Module):
  def __init__(self,input_size,output_size):
    super().__init__()
    self.linear=nn.Linear(input_size,output_size)
  def forward(self,x):
    pred=self.linear(x)
    return pred
```

```
#Creating a custom dataset and a new model
X=torch.randn(100,1)*10
y=X+torch.randn(100,1)*4
plt.plot(X.numpy(),y.numpy(),'o')
```

```
model_linear=Linear_reg(1,1)
def get_parames():
  [w,b]=model_linear.parameters()
  return (w[0][0].item(),b[0].item())

def plot_fig():
  x1=np.array([[-30],[30]])
  w,b=get_parames()
  y1=w*x1+b
  plt.plot(X.numpy(),y.numpy(),'o')
  plt.plot(x1,y1,'r')
  plt.show()
```

```
criterion=nn.MSELoss()
optimiser=torch.optim.SGD(model_linear.parameters(),lr=0.01)
```

```
epoch=100
losses=[]
for i in range(epoch):
  y_pred=model_linear.forward(X)
  loss=criterion(y_pred,y)
  losses.append(loss)
  optimiser.zero_grad()
  loss.backward()
  optimiser.step()
```

```
plt.plot(range(epoch),losses)
```

```
plot_fig()
```
