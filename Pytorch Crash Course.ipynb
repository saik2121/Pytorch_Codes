```
###Derivates in Torch
```

```
x=torch.tensor(2.0,requires_grad=True)
z=torch.tensor(1.0,requires_grad=True)
y=2*x**3+3*z**2
y.backward()
print(x.grad)
print(z.grad)
```

```
#different types of creating a tensor from np array
data=np.array([1,2,3])
t1=torch.Tensor(data)     #deep Copy float values
t2=torch.tensor(data)     #deep copy int values(Best Method)
t3=torch.as_tensor(data)  #shallow copy
t4=torch.from_numpy(data) #shallow Copy
```

```
print(type(t1))
print(type(t2))
print(t3)
print(t4)
```

#                 Squeezing and Unsqueezing
squeezing removes an axis of size 1
and unsqueezing does add a length of 1

```
t=torch.tensor([[1,2,3,4,5,6]])
t.shape
```

```
t=t.squeeze()
t.shape
```

```
t=t.unsqueeze(dim=0)
t.shape
```
